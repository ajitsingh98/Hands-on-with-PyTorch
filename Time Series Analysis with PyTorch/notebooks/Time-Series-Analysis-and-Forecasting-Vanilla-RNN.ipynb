## Data preparation

### Train/Test split

Let's divide the data into train and test set. Also decide the window size that will be used to train rnn model.

```python
all_data = df['IPG2211A2N'].values.astype(float)

window_size = 12

train_set = all_data[:-window_size]
test_set = all_data[-window_size:]

print(f"Total shape : {all_data.shape}")
print(f"Train shape : {train_set.shape}")
print(f"Test shape : {test_set.shape}")
```

### Normalize the train set

```python
scale = MinMaxScaler((-1, 1))
train_norm = scale.fit_transform(train_set.reshape(-1, 1))
train_norm.shape
```

```python
print(f"Minimum : {min(train_norm)}")
print(f"Maximum : {max(train_norm)}")
print(f"Mean : {np.mean(train_norm)}")
```

### Data preparation for RNN

```python
train_norm = torch.FloatTensor(train_norm).reshape(-1)

def input_data(seq, ws):
    out = []
    L = len(seq)
    for i in range(L-ws):
        window = seq[i:ws+i]
        label = seq[ws+i:ws+i+1]
        out.append((window, label))
    return out
```

```python
train_data = input_data(train_norm, 12)

train_data[:2]
```

We have created the train data in form of a list of tuples where each tuple represent a sequence and its label

## Define Model

We are done with data preparation now we can move ahead to create a model based on RNNs.

```python
import torch
import torch.nn as nn

class RNNForecaster(nn.Module):
    def __init__(self, input_size=1, hidden_size=50, output_size=1):
        super().__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.rnn = nn.RNN(input_size=input_size, hidden_size=hidden_size)
        self.linear = nn.Linear(hidden_size, output_size)

    def forward(self, X):
        # Initialize hidden state for each new sequence
        hidden_state = torch.zeros(1, 1, self.hidden_size)
        output, _ = self.rnn(X.view(len(X), 1, -1), hidden_state)
        preds = self.linear(output.view(len(X),-1))  # Taking the last time step's output
        return preds[-1]
```

## Instantiate the model

```python
rnnForecaster = RNNForecaster()
```

## Define the loss function and optimizer

```python
criterion = nn.MSELoss()

optimizer = torch.optim.Adam(rnnForecaster.parameters(), lr=0.001)
```

## Total number of parameters

```python
total_num = 0

for param in rnnForecaster.parameters():
    total_num += param.numel() if param.requires_grad == True else 0

print(f"Total Trainable parameters are : {total_num}")
```

## Model's training

```python
for seq, y in train_data:
    pass
```

```python
epochs = 100

import time
start_time = time.time()

for epoch in range(epochs):
    
    # extract the sequence & label from the training data
    for seq, y_train in train_data:
        # reset the parameters and hidden states
        optimizer.zero_grad()
        y_pred = rnnForecaster.forward(seq)
        loss = criterion(y_pred, y_train)
        loss.backward()
        optimizer.step()
        
    # print training result
    print(f'Epoch: {epoch+1:2} Loss: {loss.item():10.8f}')
    
print(f'\nDuration: {time.time() - start_time:.0f} seconds')
```

## Inference

We are donw with traning and now let's do inference for test points that we separated from our all data.

```python
future = 12
window_size = 12
# Add the last window of training values to the list of predictions
preds = train_norm[-window_size:].tolist()

# Set the model to evaluation mode
rnnForecaster.eval()

with torch.no_grad():
    
    for i in range(future):
        inputSeq = torch.FloatTensor(preds[-window_size:]).view(window_size, 1, -1)
        y_pred = rnnForecaster(inputSeq)
        preds.append(y_pred)

preds = preds[-window_size:]

preds
    
```

Above predicted values are normalized with MinMax method. In order to compare them with original values we need to do inverse transformation as shown below.

```python
preds_orig = scale.inverse_transform(preds)
```

## Forecasting to unknown dates

Use the trained model and predict the potential values for next 12 points i.e from `2019-02-01` to `2020-02-01`

```python
window_size = 12
future = 24

intial_points = df['IPG2211A2N'].values.astype(float)
print(len(intial_points))
#transform to min max normalization
intial_points_norm = scale.fit_transform(intial_points.reshape(-1, 1))

preds = intial_points_norm[-window_size:].reshape(-1).tolist()

for i in range(future):
    
    points_for_preds = torch.FloatTensor(preds[-window_size:])
    
    y_pred = rnnForecaster(points_for_preds)
    preds.append(y_pred.item())

#take prtedicted points
preds = preds[window_size:]

#take inverse transformation

preds_orig = scale.inverse_transform(np.array(preds).reshape(-1, 1))

preds_orig, intial_points[-future:]
```

```python
x_forecast = np.arange('2019-02-01', '2021-02-01', dtype='datetime64[M]').astype('datetime64[D]')
x_forecast
```
